```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv(r"../data/heart_disease_dataset.csv")
# Display first 5 rows
df.head()
```

```python
#Rename columns
df.rename(columns={
    'age': 'Age',
    'sex': 'Gender',
    'cp': 'Chest_Pain_Type',
    'trestbps': 'Resting_Blood_Pressure',
    'chol': 'Cholesterol',
    'fbs': 'Fasting_Blood_Sugar',
    'restecg': 'Resting_ECG',
    'thalach': 'Max_Heart_Rate',
    'exang': 'Exercise_Induced_Angina',
    'oldpeak': 'ST_Depression',
    'slope': 'ST_Slope',
    'ca': 'Major_Vessels',
    'thal': 'Thalassemia',
    'condition': 'Heart_Disease'
}, inplace=True)
df.rename(columns={'Sex': 'Gender'}, inplace=True)
```

```python
df.info()   #check Info
```

```python
#Check shape
print("Rows and Columns:", df.shape)
#CHECK DATA TYPES
print("\nData Types:")
print(df.dtypes)
```

```python
#check missing value
df.isnull().sum()
```

```python
#check duplicate rows
df.duplicated().sum()
#Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
# Numerical columns
numerical_cols = [
    'Age',
    'Resting_Blood_Pressure',
    'Cholesterol',
    'Max_Heart_Rate',
    'ST_Depression',
    'Major_Vessels'
]

# Fill missing values using median imputation
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
#handle missing values in categorical columns
categorical_cols = [
    'Gender',
    'Chest_Pain_Type',
    'Fasting_Blood_Sugar',
    'Resting_ECG',
    'Exercise_Induced_Angina',
    'ST_Slope',
    'Thalassemia',
    'Heart_Disease'
]

#USING MODE IMPUTATION
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
#Verify missing values
df.isnull().sum()
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```
