```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# The 'r' tells Python to read the single slashes perfectly without confusion
df = pd.read_csv(r"C:\Users\lenovo\Desktop\Git project\healthcare-predictive-analysis-ehr\data\synthetic_heart_disease_dataset_raw _data.csv")
df.head()

```

```python
df.info()   #check Info
```

```python
#Check shape
print("Rows and Columns:", df.shape)
#CHECK DATA TYPES
print("\nData Types:")
print(df.dtypes)
```

```python
#check missing value
df.isnull().sum()
```

```python
#check duplicate rows
df.duplicated().sum()
#Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
#handle missing values in numerical columns
print(df.dtypes)
numerical_cols = [
    'Age', 'Weight', 'Height', 'BMI',
    'Systolic_BP', 'Diastolic_BP',
    'Heart_Rate', 'Blood_Sugar_Fasting',
    'Cholesterol_Total'
]
#convert numerical column into numeric type
for col in numerical_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

#fill missing values  (USING MEDIAN IMPUTATION)
for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
#handle missing values in categorical columns
categorical_cols = [
    'Gender',
    'Smoking',
    'Alcohol_Intake',
    'Physical_Activity',
    'Diet',
    'Stress_Level'
]
#convert categorical column into category type
for col in categorical_cols:
    df[col] = df[col].astype("category")
#USING MODE IMPUTATION
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

```

```python
#Verify missing values
print(df.isnull().sum())
```

```python
#Summary statistics
print(df.describe())
```
